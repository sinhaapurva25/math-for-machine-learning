```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'movie_revenues.csv')
```

```python
df.describe()
```

```python
df.head(1)
```

```python
# df['revenue_in_million'] = df.revenue.apply(lambda x: x/1000000)
df['revenue_in_million'] = df.revenue/1000000
df.revenue_in_million.describe()
```

```python
df.head(1)
```

```python
_, mean, std, *_  = df.revenue_in_million.describe()
mean, std
```

```python
df['z_score'] = (df.revenue_in_million-mean)/std
df['z_score']
```

```python
df.head(1)
```

```python
df[df.z_score>3]
```

```python
df[df.z_score > 3].z_score
```

```python
median = df.revenue_in_million.median()
mean_absolute_dev = df.revenue_in_million.mad()
df['modified_z_score'] = 0.6745*(df.revenue_in_million-median)/mean_absolute_dev
# mean_absolute_dev
df['modified_z_score'].iloc[1]
```

```python
df[(df['modified_z_score'] >= -3.5) & (df['modified_z_score'] <= 3.5)].head(1)
```
